# ensure annotations

```python
def get_product(x: int, y:int) -> int:
    return x * y
```

```python
get_product(x=2, y=5)
```

```python
get_product(x=2, y='5')
```

```python
from ensure import ensure_annotations

@ensure_annotations
def get_prodict(x: int, y:int) -> int:
    return x * y
```

```python
get_prodict(x=2, y="5")
```
